```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'unclean_listings.csv')
df.shape
```

```python
df = df[df['header'].str.contains('Rover') == False]
df = df[df['header'].str.contains('Romeo') == False]
df = df[df['header'].str.contains('Martin') == False]
df.head()
```

```python
new_cols = df['header'].str.split(pat=' ', n=3, expand=True)
df['year'] = new_cols[0]
df['make'] = new_cols[1]
df['model'] = new_cols[2]
df['year'] = df['year'].str.replace('Sponsored','')
df = df[df['year'] != '']
df = df[['vin', 'year', 'make', 'model', 'trim', 'price', 'location', 'mileage', 'exterior_color', 'interior_color', 'fuel_type', 'mpg', 'transmission', 'drivetrain']]
df.head()
```

```python
df['trim'] = df['trim'].str.replace('4WD', '')
df['trim'] = df['trim'].str.replace('AWD', '')
df['trim'] = df['trim'].str.replace('FWD', '')
df['trim'] = df['trim'].str.replace('RWD', '')
df['trim'] = df['trim'].str.replace('2WD', '')
df['trim'] = df['trim'].str.replace('Automatic', '')
df['trim'] = df['trim'].str.replace('Auto', '')
df['trim'] = df['trim'].str.replace('Manual', '')
df['trim'] = df['trim'].str.replace('CVT', '')
df.head()
```

```python
new_cols = df['price'].str.split(pat='$', n=2, expand=True)
df['price'] = new_cols[1]

df['price'] = df['price'].str.replace(',', '')
df.head()
```

```python
new_cols = df['location'].str.split(pat=',', n=2, expand=True)
df['city'] = new_cols[0]
df['state'] = new_cols[1]
df.drop(['location'], axis=1, inplace=True)
df.head()
```

```python
df['mileage'] = df['mileage'].str.replace(',', '')
df['mileage'] = df['mileage'].str.replace('miles', '')
df.head()
```

```python
df['exterior_color'] = df['exterior_color'].str.replace('Exterior: ', '')
df.head()
```

```python
df['interior_color'] = df['interior_color'].str.replace('Interior: ', '')
df.head()
```

```python
df['fuel_type'] = df['fuel_type'].str.replace('Fuel Type: ', '')
df.head()
```

```python
new_cols = df['mpg'].str.split(pat=' ', n=5, expand=True)
df['city_mpg'] = new_cols[1]
df['hwy_mpg'] = new_cols[4]
df.drop(['mpg'], axis=1, inplace=True)

df = df[['vin', 'year', 'make', 'model', 'trim', 'price', 'mileage', 'transmission', 'drivetrain', 'fuel_type', 'city_mpg', 'hwy_mpg', 'exterior_color', 'interior_color', 'city', 'state']]
df.head()
```

```python
df['transmission'] = df['transmission'].str.replace(' Transmission', '')
df.head()
```

```python
df['vin'] = df['vin'].str.strip()
df['year'] = df['year'].str.strip()
df['make'] = df['make'].str.strip()
df['model'] = df['model'].str.strip()
df['transmission'] = df['transmission'].str.strip()
df['drivetrain'] = df['drivetrain'].str.strip()
df['fuel_type'] = df['fuel_type'].str.strip()
df['exterior_color'] = df['exterior_color'].str.strip()
df['interior_color'] = df['interior_color'].str.strip()
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.dropna(subset=['city_mpg'], inplace=True)
df = df[df['city_mpg'] != 'N/A']
```

```python
df['year'] = df['year'].astype(int)
df['price'] = df['price'].astype(int)
df['mileage'] = df['mileage'].astype(int)
df['city_mpg'] = df['city_mpg'].astype(int)
df.info()
```
